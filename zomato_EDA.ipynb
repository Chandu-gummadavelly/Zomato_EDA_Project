```python
df2=pd.read_excel('Country-Code.xlsx')
```

```python
df2.head()
```

```python
# we have combined two tables
final_df=pd.merge(df,df2,on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
final_df.dtypes
```

```python
final_df.Country.value_counts()
```

 maximum transactions from india

```python
country_names= final_df.Country.value_counts().index
```

```python
country_names
```

```python
country_values=final_df.Country.value_counts().values
```

```python
plt.pie(country_values,labels=country_names,autopct='%1.2f%%')
plt.show()
```

```python
# here we are plotting top3 countries
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
plt.show()
```

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text'])
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

## Observations
1. if rating is between 4.5-4.9 --> Excellent
2. if rating is between 4.0-4.4 --> Very Good
3. if rating is between 3.5-3.9 --> Good
4. if rating is between 2.5-3.4 --> Average
5. if rating is between 1.8-2.4 --> poor
6. if rating is 0.0 --> not rated

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
```

```python
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings,color='pink')
plt.show()
```

```python
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings,hue='Rating color',palette=['Blue','Red','Orange','Yellow','Green','Green'])
plt.show()
```

## Observations
1. not rated count is high
2. and most of the ratings are between 2.5-3.4

```python
# count plot
sns.countplot(x='Rating color',data=ratings,palette=['Blue','Red','Orange','Yellow','Green','Green'])
plt.show()
```
